Transform iris data into 3 dimensions
and plot a 3d chart with transformed dimensions and colour each data
point with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
iris=datasets.load_iris()
```

```python
x=PCA(n_components=3).fit_transform(iris.data)
y=iris.target
```

```python
fig=plt.figure(1,figsize=(8,6))
axis=Axes3D(fig,elev=-150,azim=110)
axis.scatter(x[:,0],x[:,1],x[:,2],c=y,cmap=plt.cm.prism_r)
axis.set_title("First three PCA directions")
axis.set_xlabel("Principle component 1")
axis.set_ylabel("Principle component 2")
axis.set_zlabel("Principle component 3")
plt.show()
```
